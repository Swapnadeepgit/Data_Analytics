TASK NO.2 - Unemployment is measured by unemployment rate which is the no. of people who are unemployed as a percentage of the total labour force . We have seen sharp increase in the unemployment rate during the Covid-19 , so analysing the unemployment rate can be a good data science project





```python
#importing all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing the datasets
data=pd.read_csv('/content/Unemployment in India.csv')
data=pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(17,6))
fig, ax=plt.subplots()

diag = plt.bar(data['Region'], data[' Estimated Unemployment Rate (%)'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=60 ,ha='right')
plt.title("Indian Unemployment")
plt.xlabel("States")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
#creation of pair plot of datasets
sns.pairplot(data, hue='Region.1',palette='Dark2')
```

```python
sns.histplot(data[' Estimated Unemployment Rate (%)'])
```

```python
sns.histplot(data[' Estimated Employed'])
```

```python
#estimated employment rate in accordance with the region
sns.histplot(x= ' Estimated Employed', data=data, hue='Region.1')
```

```python
#impact of covid-19 on Unemployement rate over time
fig = px.line(data, x=' Date', y=' Estimated Unemployment Rate (%)', color='Region', title='Unemployment Rate Over Time',labels={'Estimated Unemployment Rate (%)':'Unemployment Rate'}, template='plotly_dark')
covid_events = {'start_date': '2020-03-01', 'end_date': '2020-03-31', 'event':'Peak of Pandemic'}
fig.add_vrect(x0=covid_events['start_date'], x1=covid_events['end_date'], fillcolor='red', opacity=0.2, line_width=0)
fig.show()
```

```python
#impact of unemployement rate over times in accordance with the states with animation
fig = px.bar(data, x='Region', y=' Estimated Unemployment Rate (%)', color='Region',
             animation_frame=' Date', title='Unemployment Rate Over Time',
             labels={'Estimated Unemployment Rate (%)':'Unemployment Rate'},
             template='plotly',
             color_discrete_sequence = px.colors.sequential.Viridis,
             category_orders={'Region':['Region1','Region2','Region3',]})
fig.update_layout(xaxis={'categoryorder':'total descending'},
           xaxis_title='Region', yaxis_title='Unemployment Rate (%)',
                  barmode='stack',
                  showlegend=False)
fig.add_annotation(text='Impact of Unemployment Over Time',
                  xref='paper',yref='paper',
                  showarrow=False,
                  font=dict(size=16, color='black'))
fig.show()
```
